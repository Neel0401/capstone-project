```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_html('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M',header=0)
```

```python
df1=df[0]
```

```python
df1.head()
```

```python
df1.rename(columns={'Post Code':'Postalcode'},inplace=True)
```

```python
df1.head()
```

```python
df2=df1[df1['Borough']!= 'Not assigned']
```

```python
df2.head()
```

```python
df3=pd.read_csv('Geospatial_Coordinates.csv')
```

```python
df3.head()
```

```python
df_toronto=pd.merge(df2,df3,how='left',left_on='Postalcode',right_on='Postal Code')
```

```python
df_toronto.head()
```

```python
df_toronto.drop('Postal Code',axis=1,inplace=True)
```

```python
df_toronto.head()
```

```python
address = 'Toronto, ON, Canada'
geolocator = Nominatim(user_agent="to_explorer")
location = geolocator.geocode(address)
latitude = location.latitude
longitude = location.longitude
print('coordinate of Toronto, ON, Canada are {}, {}.'.format(latitude, longitude))
```

```python
map_toronto = folium.Map(location=[latitude, longitude], zoom_start=10)
for lat, lng, borough, neighborhood in zip(df_toronto['Latitude'], df_toronto['Longitude'], df_toronto['Borough'], df_toronto['Neighborhood']):
    label = '{}, {}'.format(neighborhood, borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(map_toronto)  
    
map_toronto
```
